```python
import pandas as pd
```

```python
dados_emprestimos = pd.read_parquet("Dados_completos.parquet")
dados_emprestimos.head()
```

```python
CDU_list = []
for loc in dados_emprestimos['localizacao']:
    if (loc < 100):
        CDU_list.append("Gen_Ciê_Con")
    elif (loc < 200):
        CDU_list.append("Filo_Psico")
    elif (loc < 300):
        CDU_list.append("Religião")
    elif (loc < 400):
        CDU_list.append("Ciências Sociais")
    elif (loc < 500):
        CDU_list.append("Não Ocupada")
    elif (loc < 600):
        CDU_list.append("Ciências Naturais")
    elif (loc < 700):
        CDU_list.append("Ciências Aplicadas")
    elif (loc < 800):
        CDU_list.append("Belas Artes")
    elif (loc < 900):
        CDU_list.append("Línguas")
    else:
        CDU_list.append("Geografia História")

dados_emprestimos["CDU_geral"] = CDU_list
dados_emprestimos.head()
```

```python
dados_emprestimos.drop(columns=['registro_sistema'], inplace=True)
```

```python
dados_emprestimos['matricula_ou_siape'] = dados_emprestimos['matricula_ou_siape'].astype('string')
```
