```python
import pandas as pd
```

```python
dados_emprestimos = pd.read_parquet("Dados Concatenados/emprestimos.parquet")
dados_exemplares = pd.read_parquet("https://github.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/blob/main/Dia_1-Importando_dados/Datasets/dados_exemplares.parquet?raw=true")
```

```python
dados_emprestimos = dados_emprestimos.drop_duplicates(ignore_index=True)
del dados_emprestimos["Unnamed: 0"]
dados_emprestimos.head()
```

```python
dados_emprestimos.value_counts()
```

```python
dados_juntos = dados_emprestimos.merge(dados_exemplares)
dados_juntos
```

```python
dados_juntos.to_parquet("Dados Concatenados/Dados_completos.parquet")
```
