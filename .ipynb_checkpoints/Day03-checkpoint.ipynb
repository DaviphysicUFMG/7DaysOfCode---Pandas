```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dados_emprestimos = pd.read_parquet("dados_completos_dia2.parquet")
dados_emprestimos.head()
```

```python
dados_emprestimos['id_emprestimo'].value_counts()
```

```python
dados_emprestimos['data_devolucao'] = pd.to_datetime(dados_emprestimos['data_devolucao'])
dados_emprestimos['data_emprestimo'] = pd.to_datetime(dados_emprestimos['data_emprestimo'])
dados_emprestimos['data_renovacao'] = pd.to_datetime(dados_emprestimos['data_renovacao'])
dados_emprestimos.head()
```

```python
data_emprestimos = pd.DataFrame(dados_emprestimos['data_emprestimo'].value_counts()).reset_index()
data_emprestimos.columns = ['data', 'quantidade']
data_emprestimos
```

```python
ano_emprestimos = data_emprestimos.groupby(by=data_emprestimos.data.dt.year).sum()
ano_emprestimos.index.name = 'ano'
ano_emprestimos
```

```python
sns.set_theme(
    context='notebook',
    style='darkgrid',
    palette='deep',
    font_scale=1.2,
    rc={"figure.figsize":(10,5)}
)

ax = sns.lineplot(data=ano_emprestimos, x='ano', y='quantidade')
ax.set(xlabel='Ano', ylabel=None)
ax.tick_params(axis='x', rotation=45)
ax.set_title('Quantidade de emprestimos por Ano', size=25, loc='center', weight='bold')
```

```python
mes_emprestimos = data_emprestimos.groupby(by=data_emprestimos.data.dt.month).sum()
mes_emprestimos.index.name = 'mes'

dic_mes = {
    1: 'Jan',
    2: 'Fev',
    3: 'Mar',
    4: 'Abr',
    5: 'Mai',
    6: 'Jun',
    7: 'Jul',
    8: 'Ago',
    9: 'Set',
    10: 'Out',
    11: 'Nov',
    12: 'Dez'
}

mes_emprestimos.index = mes_emprestimos.index.map(dic_mes)
mes_emprestimos
```

```python
sns.set_theme(
    context='notebook',
    style='darkgrid',
    palette='deep',
    font_scale=1.2,
    rc={"figure.figsize":(10,5)}
)

ax = sns.lineplot(data=mes_emprestimos, x='mes', y='quantidade')
ax.set(xlabel='Mês', ylabel=None)
ax.tick_params(axis='x', rotation=45)
ax.set_title('Quantidade de emprestimos por Mês', size=25, loc='center', weight='bold')
```
